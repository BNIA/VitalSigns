# Welcome

This script will prepare our VitalSignsXl Doc for upload. 

1. It uses the meta data table to replace the datasheets column headers with a shortname
2. Then empties columns that should not have their data uploaded yet according to section
3. Then it Deletes placeholder columns for our wishlist indicators

```
xlsheet = '19'
year = '18'
```

```
import pandas as pd
# df = pd.read_excel(open('VS19_Indicator_Data_Form.xlsm', 'rb'), sheet_name="Vital Signs 19 Indicators", header=None, names=['vsid', 'xlsmid', 'longname','shortname', 'UPLOAD_WP',])  
df = pd.read_excel(open('VS'+xlsheet+'_Indicator_Data_Form.xlsm', 'rb'), sheet_name="Vital Signs "+xlsheet+" Indicators")
df = df[['ShortName', 'Section']]
df.ShortName = df.ShortName.astype(str).str[:-2]
df.loc[-1] = ['CSA2010', 'Section'] 
df.index = df.index + 1
df.sort_index(inplace=True) 
display( df.head(1) ) 
```

Replace Column Headers of our data with the ShortNames

```
import pandas as pd
vsdf = pd.read_excel(open('VS'+xlsheet+'_Indicator_Data_Form.xlsm', 'rb'), sheet_name='VS'+year, header=1)
vsdf = vsdf.drop(columns='Unnamed: 0')
vsdf.columns = df["ShortName"].values
display( vsdf.columns )
display( vsdf.head(1) )
```

```
vsdf.head(1)
```

Now Remove Data from Columns Not on Our Publishing List.

```
keep = ['Section', 'Crime and Safety','Census Demographics',
       'Workforce and Economic Development', 'Sustainability', 
       'Housing and Community Development', 'Children and Family Health', 
        'Arts and Culture', 'Education and Youth']
rm = []

# For Each Column
for currentColumn in vsdf.columns:
  # Get the Section
  indSection = df[ df['ShortName'] == currentColumn ]['Section'].values[0]
  # Delete Records if Not In Keep List
  if ( not indSection in set( keep ) ):
    vsdf[currentColumn] = ''

# Delete columns if on wishlist.
vsdf = vsdf[ df[ df['ShortName'] != 'n' ].ShortName.values ]

vsdf = vsdf.add_suffix(year) 
vsdf.rename(columns = {f"CSA2010{year}": "CSA2010"}, inplace=True)
# vsdf.fillna(-1, inplace=True)
vsdf = vsdf.round(2)

vsdf.head()
```

```
vsdf.to_csv('export'+year+'.csv', index=False)
```
